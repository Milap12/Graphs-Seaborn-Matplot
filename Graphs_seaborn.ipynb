```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### There are 17 built-in datasets available in Seaborn



```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
#flights = sns.load_dataset('flights')
```

```python
tips.head()
```

### Calling sns.set() without any arguments sets Seaborn to use its default aesthetics for various aspects of plots, including:
Color palette: "darkgrid" theme by default (adjusts based on the context)
Font size: Slightly larger than Matplotlib's defaults
Style elements: Grid lines, spines, etc.

```python
sns.set()
sns.swarmplot(x='species', y='petal_length', data=iris)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.set()
from seaborn import color_palette
palette = color_palette("Set1")
sns.barplot(x='day', y='total_bill',palette=palette, data=tips)
```

```python
sns.set_style('ticks')
sns.boxplot(x='day', y='total_bill',palette=palette, data=tips)
```

```python
sns.set_context("paper")
sns.set_style("whitegrid")
sns.boxplot(x = "day", y = "total_bill", palette=palette,data = tips)
```

```python
# Creating a randomly distributed dataset:
x = np.random.normal(size=100)
sns.distplot(x,bins=20,kde=True)  #kernel density estimate
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips, color='r')
```

```python
# Creating Strip plot for day-wise revenue:
sns.stripplot(x="day", y="total_bill", data=tips, color="g")
```

```python
sns.set_style('whitegrid')

# Creating Strip plot for day-wise revenue:
sns.swarmplot(x="day", y="total_bill", data=tips, palette="viridis")
```

```python
# Basic lmplot visualization:
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, markers=["o", "x"], palette="Set1")
# legend=True
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, lowess=True)
```
